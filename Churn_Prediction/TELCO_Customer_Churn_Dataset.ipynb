# About This File

This dataset has details of 7,043 telecom customers, like their personal info, services they use, and payment details. It's mainly used to understand why customers leave and to predict who might leave in the future.

# Problem Statement

In today’s telecom market, keeping current customers is just as important as getting new ones. Many telecom companies lose customers (called churn) when they stop using their services and move to other providers. This results in losing money and spending more to find new customers.

The telecom company in this dataset is facing the same issue and wants to find out why some customers are leaving. People may leave because of poor service, high prices, better deals from other companies, or lack of support. If the company can spot these reasons early, they can make improvements and try to keep their customers happy.

To help with this, the company has collected data from 7,043 customers, including their personal info, the services they use, how they pay, and whether they left the company or not.

The goal is to understand the reasons behind customer churn by analyzing the data and to build a model that can predict whether a customer is likely to leave. This will help the company take early action to retain customers and improve customer satisfaction.

# Features

# Column Description

Here’s a description of the features in your dataset:

**customerID** – A unique number or code given to each customer.

**gender** – Tells if the customer is male or female.

**SeniorCitizen** – Shows if the customer is a senior citizen (1 means Yes, 0 means No).

**Partner** – Says if the customer has a husband, wife, or partner (Yes/No).

**Dependents** – Tells if the customer has kids or others depending on them (Yes/No).

**tenure** – How many months the customer has been using the company’s services.

**PhoneService** – Shows if the customer has a phone connection (Yes/No).

**MultipleLines**– Tells if the customer has more than one phone line.

**InternetService**– Type of internet the customer uses (DSL, Fiber optic, or None).

**OnlineSecurity** – Says if the customer has a security service for the internet.

**OnlineBackup** – Shows if the customer has a service to back up their data online.

**DeviceProtection** – Tells if the customer has protection for their devices.

**TechSupport** – Says if the customer gets technical help or support.

**StreamingTV** – Shows if the customer watches TV online.

**StreamingMovies** – Tells if the customer watches movies online.

**Contract** – Type of agreement with the company (monthly, 1 year, or 2 years).

**PaperlessBilling** – Says if the customer gets bills online instead of on paper.

**PaymentMethod** – How the customer pays (like with a credit card or bank).

**MonthlyCharges** – The amount the customer pays every month.

**TotalCharges** – The total money the customer has paid so far.

**Churn** – Tells if the customer has stopped using the service (Yes means they left).

# Target Variable

It shows whether a customer has left the company or not:

Yes – The customer has stopped using the service.

No – The customer is still using the service.

# KFold

```python
from sklearn.model_selection import KFold
kf=KFold(n_splits=5,shuffle=True,random_state=42)
model=LogisticRegression()
```

```python
best_accuracy=0
best_fold=1
```

```python
best_x_train=None
best_y_train=None
best_x_test=None
best_y_test=None
```

```python
from sklearn.metrics import accuracy_score
for fold,(train_index,test_index) in enumerate(kf.split(x),1):
  xtrain,xtest=x.iloc[train_index],x.iloc[test_index]
  ytrain,ytest=y.iloc[train_index],y.iloc[test_index]

  model.fit(xtrain,ytrain)

  ypred=model.predict(xtest)
  acc=accuracy_score(ytest,ypred)
  print("_____________________________")
  print(f'fold{fold} has accuracy:{acc}')

  #update best fold
  if acc>best_accuracy:
    best_accuracy=acc
    best_fold=fold
    best_x_train=xtrain
    best_x_test=xtest
    best_y_train=ytrain
    best_y_test=ytest
    best_y_pred=ypred
#output the best
print(f"\n Best Fold:{best_fold} with accuracy:{best_accuracy}")
```
